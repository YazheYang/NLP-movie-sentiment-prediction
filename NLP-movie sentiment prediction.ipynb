```python
from bs4 import BeautifulSoup 
ex1 = BeautifulSoup (train['review'][0],"lxml")
print train["review"][0]
print ex1.get_text()
```

```python
from nltk.corpus import stopwords
sw = set(stopwords.words('english'))
len(sw)
```

```python
import re
letter_only = re.sub("[^a-zA-Z]", " ", ex1.get_text())
print letter_only
```

```python

#tokenization
lower_case = letter_only.lower()
words = lower_case.split()
```

```python
words = [w for w in words if w not in sw]
print words
```

```python
from nltk.stem.snowball import SnowballStemmer
new_words = []
stemmer = SnowballStemmer("english")
for i in words:
    new_words.append(stemmer.stem(i))
print new_words    
```

```python
def review_to_words( raw_review ):
    new_words = []
    review_text = BeautifulSoup(raw_review, "lxml").get_text() 
    letter_only = re.sub("[^a-zA-Z]", " ", review_text) 
    words = letter_only.lower().split()  
    words = [w for w in words if w not in sw]
    for i in words:
        new_words.append(stemmer.stem(i))
    final_words = " ".join(new_words)
    return final_words
```

```python
num_reviews = len(train["review"])
clean_train_reviews = []
for i in range(num_reviews):
    if( (i+1)%1000 == 0 ):
        print "Review %d of %d\n" % ( i+1, num_reviews ) 
    clean_train_reviews.append(review_to_words( train['review'][i] ))
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5000)
###bags_of_words = vectorizer.fit(clean_train_reviews)
###train_data_features = bags_of_words.transform(clean_train_reviews)
train_data_features = vectorizer.fit_transform(clean_train_reviews)

print train_data_features[1]
```

```python
train_data_features = train_data_features.toarray()
print train_data_features.shape
print train_data_features[1]
```

```python
vocal = vectorizer.get_feature_names()
```

```python
import numpy as np
words_occur_times = np.sum(train_data_features,axis=0)
for i, j in zip(words_occur_times, vocal):
    
    print j, i
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(train_data_features, train['sentiment'])
```

```python
test = pd.read_csv("testData.tsv",delimiter="\t" )
num_test_reviews = len(test["review"])
clean_test_reviews = []
for i in range(num_test_reviews):
    if( (i+1)%1000 == 0 ):
        print "Review %d of %d\n" % ( i+1, num_test_reviews ) 
    clean_test_reviews.append(review_to_words( test['review'][i] ))
```

```python
test_data_features = vectorizer.transform(clean_test_reviews).toarray()
pred = clf.predict(test_data_features)

output = pd.DataFrame({'id':test["id"], "sentiment":pred})
output.to_csv('movie_sentiment_prediction', index=False, quoting=3)
```
